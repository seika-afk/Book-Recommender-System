```python
#mergin ->on basis of isbn
ratings_with_name=ratings.merge(books,on="ISBN")

ratings_with_name
```

```python
num_rating_df =ratings_with_name.groupby('Book-Title').count()['Book-Rating'].reset_index()
# .reset_index->gives df from dictt
```

```python
num_rating_df.rename(columns={'Book-Rating':"num_ratings"},inplace=True)
```

```python
num_rating_df
```

```python
avg_rating_df = ratings_with_name.groupby('Book-Title')['Book-Rating'].mean().reset_index()
avg_rating_df.rename(columns={"Book-Rating": "avg_rating"}, inplace=True)
```

```python
avg_rating_df
```

```python
popularity_df = num_rating_df.merge(avg_rating_df,on="Book-Title")
```

```python
popularity_df
```

```python
popularity_df=popularity_df[popularity_df["num_ratings"]>=250].sort_values("avg_rating",ascending=False).head(50)
```

```python
popularity_df.shape
```

```python
#now we need author name,image
```

# Collaborative Filtering Based Recommender System

```python
x=ratings_with_name.groupby('User-ID').count()['Book-Rating'] >200
# x[x] #just getting true values

padheUsers=x[x].index

# users with count of voted >200
```

```python
filtered_rating=ratings_with_name[ratings_with_name['User-ID'].isin(padheUsers)]
```

```python
y=filtered_rating.groupby('Book-Title').count()['Book-Rating']>=50
#books having rating >50
famous_books=y[y].index
```

```python
final_ratings=filtered_rating[filtered_rating['Book-Title'].isin(famous_books)]
```

```python
final_ratings.duplicated().sum()
```

```python
#now creating table where columns are users and rows are books and values are
#ratings by users
```

```python
pt=final_ratings.pivot_table(index='Book-Title',columns='User-ID',values='Book-Rating')
```

```python
pt.fillna(0,inplace=True)
```

```python
pt
```

```python
#model building
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity_scores=cosine_similarity(pt)
```

```python
similarity_scores.shape
#1 book with every book,their cosine similarity
```

```python
#making a function -> giving input 1 book and it gives output 5 similar books
```
